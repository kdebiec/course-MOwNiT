# Zadanie 4

Przykładem takiego algorytmu może być wyznaczanie e^x

Podpunkt 1)

```julia
function powerofe(power::Float64)
    e = Float64(1)
    e += power
    for i=2:20
        e += (power^i)/factorial(i)
    end
    return e
end
```

```julia
powerofe(-5.5)
```

Podpunkt 2)

W powyższym algorytmie zwiększenie błędu jest spowodowane utratą cyfr znaczących w momencie dodawania i odejmowania coraz mniejszych liczb. W poniższym algorytmie operacja ta została zmieniona i z wartości ujemnych bierzemy wartość absolutną dalej dodając do liczby e, na końcu jeśli wykładnik był ujemny w wyniku wyliczoną liczbą dzielimy 1

Podpunkt 3)

```julia
function powerofe2(power::Float64)
    isNegative = Bool(power < 0)
    power = sign(power)*power
    e = Float64(1)
    e += power
    for i=2:20
        e += (power^i)/factorial(i)
    end
    
    if isNegative
        e = Float64(1/e)
    end
    
    return e
end
```

```julia
powerofe2(-5.5)
```
